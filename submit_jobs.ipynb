```python
import subprocess, os, time
from termcolor import cprint

MAX_JOBS = 1950
MAX_SIZE_GB = 60.0

def checkNumJobs():
    subproc = subprocess.Popen("ssh deigo 'squeue'", shell=True, stdout=subprocess.PIPE)
    ret = subproc.stdout.read().decode('UTF-8').split('\n')[1:-1]
    n_jobs_now = len(ret)
    cprint(f'Number of jobs running: {n_jobs_now}', color='red', attrs=['bold'])
    return n_jobs_now
```

```python
n_jobs_now = checkNumJobs()

for i in range(2017, 10000):
    time.sleep(3)
    
    # after a pause, check how many jobs are running AND the size of the directory:
    n_jobs_now = checkNumJobs()
    this_folder_size_GB = sum(os.path.getsize(f) for f in os.listdir('.') if os.path.isfile(f))/1024**3
    
    while ((n_jobs_now > MAX_JOBS) | (this_folder_size_GB > MAX_SIZE_GB)):
        time.sleep(30)
        n_jobs_now = checkNumJobs()
        this_folder_size_GB = sum(os.path.getsize(f) for f in os.listdir('.') if os.path.isfile(f))/1024**3
    
    # if n_jobs_now < 1997, send command to cd to the WD and start a job:
    commands = f'"cd /home/r/roman-koshkin/H2/GRID_withFD/ && sbatch start_job.sh {i}"'
    os.system(f"ssh deigo {commands}")
    print(f'Size of the folder: {this_folder_size_GB:.2f} GB. Job with cell_id {i} submitted')
        
```
